### 照片放進yolov8檢測

```python
from ultralytics import YOLO

# 初始化 YOLOv8 nano 版本
model = YOLO("yolov8n.pt")

# 執行物件偵測
results = model.predict(source='./image/image01.jpg', save=True, conf=0.5, save_txt=True)

# 顯示檢測結果
for result in results:
    result.show()
```

### 影片放進yolov8檢測

```python
from ultralytics import YOLO

# 初始化 YOLOv8 模型
model = YOLO("yolov8n.pt")

# 執行影片物件偵測
results = model.predict(source='./image/demo.mp4', save=True, conf=0.5)
print("影片處理完成，結果已儲存。")
```
