```python
import pandas as pd
import matplotlib.pyplot as plt

# 載入模型 B 與 C 的資料（請換成你自己 CSV 路徑）
model_b = pd.read_csv("model_b.csv")
model_c = pd.read_csv("model_c.csv")

epochs = range(1, len(model_b) + 1)

# 設定圖大小與字體
plt.rcParams['font.size'] = 10
plt.figure(figsize=(18, 5))

# ===== 1. Generator Loss 趨勢圖 =====
plt.subplot(1, 3, 1)
plt.plot(epochs, model_b["loss G GAN valid"], label="Model B - GAN Loss")
plt.plot(epochs, model_b["loss G L1 valid"], label="Model B - L1 Loss")
plt.plot(epochs, model_c["loss G GAN valid"], label="Model C - GAN Loss", linestyle="--")
plt.plot(epochs, model_c["loss G L1 valid"], label="Model C - L1 Loss", linestyle="--")
plt.title("Generator Loss Comparison (Valid)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# ===== 2. 判別器準確率（真假圖） =====
plt.subplot(1, 3, 2)
plt.plot(epochs, model_b["predict real accuracy valid"], label="Model B - Real Accuracy")
plt.plot(epochs, model_b["predict fake accuracy valid"], label="Model B - Fake Accuracy")
plt.plot(epochs, model_c["predict real accuracy valid"], label="Model C - Real Accuracy", linestyle="--")
plt.plot(epochs, model_c["predict fake accuracy valid"], label="Model C - Fake Accuracy", linestyle="--")
plt.title("Discriminator Accuracy (Valid)")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

# ===== 3. L1 Loss 收斂圖（Valid）=====
plt.subplot(1, 3, 3)
plt.plot(epochs, model_b["loss G L1 valid"], label="Model B")
plt.plot(epochs, model_c["loss G L1 valid"], label="Model C")
plt.title("L1 Loss (Valid) Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("L1 Loss")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取資料（請改成你實際的CSV檔路徑）
model_a = pd.read_csv("model_a.csv")
model_b = pd.read_csv("model_b.csv")

epochs_a = range(1, len(model_a) + 1)
epochs_b = range(1, len(model_b) + 1)

plt.rcParams['font.size'] = 10
plt.figure(figsize=(18, 5))

# 1️⃣ Generator Loss 比較（GAN + L1）
plt.subplot(1, 3, 1)
plt.plot(epochs_a, model_a["loss G GAN valid"], label="Model A - GAN Loss", color="blue")
plt.plot(epochs_a, model_a["loss G L1 valid"], label="Model A - L1 Loss", color="blue", linestyle="--")
plt.plot(epochs_b, model_b["loss G GAN valid"], label="Model B - GAN Loss", color="orange")
plt.plot(epochs_b, model_b["loss G L1 valid"], label="Model B - L1 Loss", color="orange", linestyle="--")
plt.title("Generator Loss Comparison")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# 2️⃣ Discriminator Accuracy 比較
plt.subplot(1, 3, 2)
plt.plot(epochs_a, model_a["predict real accuracy valid"], label="Model A - Real", color="blue")
plt.plot(epochs_a, model_a["predict fake accuracy valid"], label="Model A - Fake", color="blue", linestyle="--")
plt.plot(epochs_b, model_b["predict real accuracy valid"], label="Model B - Real", color="orange")
plt.plot(epochs_b, model_b["predict fake accuracy valid"], label="Model B - Fake", color="orange", linestyle="--")
plt.title("Discriminator Accuracy Comparison")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

# 3️⃣ L1 Loss 單獨比較
plt.subplot(1, 3, 3)
plt.plot(epochs_a, model_a["loss G L1 valid"], label="Model A (type93)", color="blue")
plt.plot(epochs_b, model_b["loss G L1 valid"], label="Model B (type83)", color="orange")
plt.title("L1 Loss Comparison (Valid)")
plt.xlabel("Epoch")
plt.ylabel("L1 Loss")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
