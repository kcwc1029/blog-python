```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('./data set/50_Startups.csv')

dataset.head()
```

```python
# 拆分 X 和 y
X = dataset.iloc[:, :-1]  # 選取除最後一列以外的所有列
y = dataset.iloc[:, -1] # 選取最後一列作為y
```

- Encoding Categorical Data（編碼類別數據）的目的是將文字類別數據轉換為數字格式
- One-Hot Encoding：為每個類別創建一列，當某行屬於該類別時，對應列的值為 1，否則為 0。

```python
from sklearn.compose import ColumnTransformer # 用於對數據集中的特定列執行變換，並同時保留其他列。
from sklearn.preprocessing import OneHotEncoder # One-Hot Encoding 
ct = ColumnTransformer(
    # 定義了需要應用的轉換操作
    transformers=[(
        'encoder', # 轉換器名稱
        OneHotEncoder(), # 使用 One-Hot 編碼
        [3] # 需要應用編碼的列索引
        )], 
    remainder='passthrough' # 對於未經處理的列，保留它們的原始值
    )

# 執行編碼與變換
X = np.array(ct.fit_transform(X))
```

```python
X
```

## Splitting dataset

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.8, test_size = 0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## training

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting results

```python
y_pred = regressor.predict(X_test)
results_df = pd.DataFrame({"y_pred":y_pred})

results_df
```
